```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data1.csv")
```

```python
data
```

```python
plt.scatter(data['Areas'] , data['Rents'] , label = 'Data' , color = 'blue' , marker = 'o')
plt.xlabel('Areas')
plt.ylabel('Rents')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3).fit(data[['Areas' , 'Rents']])
```

```python
plt.scatter(data['Areas'] , data['Rents'] , c = kmeans.labels_.astype(float), alpha=0.5)
plt.xlabel('Areas')
plt.ylabel('Rents')
x_left , x_right = plt.xlim()
y_left , y_right = plt.ylim()
plt.show()
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
plt.scatter(centroids[: , 0] , centroids[: , 1] , c = 'r')
plt.xlabel('Areas')
plt.ylabel('Rents')
plt.xlim(x_left , x_right)
plt.ylim(y_left , y_right)
plt.show()
```

```python
plt.scatter(data['Areas'] , data['Rents'] , c = kmeans.labels_.astype(float), alpha=0.5)
plt.scatter(centroids[: , 0] , centroids[: , 1] , c = 'r' , label = 'Centroids')
plt.xlabel('Areas')
plt.ylabel('Rents')
plt.legend(loc = 'upper right')
plt.show()
```

```python
plt.scatter(data['Areas'] , data['Rents'] , c = kmeans.labels_.astype(float), alpha=0.5)
plt.scatter(centroids[: , 0] , centroids[: , 1] , c = 'r' , label = 'Centroids')
plt.xlabel('Areas')
plt.ylabel('Rents')
plt.legend(loc = 'upper right')
plt.xlim(0 , 2000)
plt.ylim(0 , 40000)
plt.show()
```

```python
plt.scatter(data['Areas'] , data['Rents'] , c = kmeans.labels_.astype(float), alpha=0.5)
plt.scatter(centroids[: , 0] , centroids[: , 1] , c = 'r' , label = 'Centroids')
plt.xlabel('Areas')
plt.ylabel('Rents')
plt.legend(loc = 'upper right')
plt.xlim(1000 , 5500)
plt.ylim(43000 , 100000)
plt.show()
```
